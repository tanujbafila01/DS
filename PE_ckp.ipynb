```python
# hackerrank problem solving array: dynamic array

def dynamicArray(n, queries):
    # initialize lastAnswer to 0
    lastAnswer = 0
    return_val = []
    # create list of empty sequences
    seqList = [[] for x in range(n)]
    for query in queries:
        # expand the list to variables so that we don't have to get values based on index again and again
        queryType, x, y = query        
        if queryType == 1:
            # find seq at index ((x^lastAnswer)%n) and append y to it
            seqList[(x^lastAnswer)%n].append(y)
        if queryType == 2 :
            seq = seqList[(x^lastAnswer)%n]
            lastAnswer = seq[y%len(seq)]
            return_val.append(lastAnswer)
    return return_val
```

```python
# hackerrank delete duplicate values froma sorted linked list
def deleteDuplicateData(head):
    """
    :param head: head to a sorted linked list containing duplicated data
    """
    if not head: return None
    prev = None
    cur = head
    while cur.next:
        prev = cur
        cur = cur.next
        if prev.data==cur.data:
            while cur.next and prev.data==cur.data:
                cur = cur.next
            prev.next = cur
        if not cur.next and prev.data==cur.data:
            prev.next=None
            
    return head

# other people's solution
# recursive approach
def deleteDuplicateDataRecursion(head):
    if not head : return None
    nxt = head.next
    while nxt and head.data==nxt.data:
        nxt = nxt.next
    head.next = deleteDuplicateDataRecursion(nxt)
    return head

# simpler approac without recursion
def removeDuplicates(head):
    if head:
        node = head
        while node.next:
            if node.data == node.next.data:
                node.next = node.next.next
            else:
                node = node.next
    return head
```

```python
# hackerrank detect cycle in a linked list
# this method has high space requirement, in worst case, it will store all nodes as a set
def has_cycle(head):
    if not head: return None
    nodes = set()
    while head:
        if head in nodes:
            return 1
        if head not in nodes:
            nodes.add(head)
        head = head.next
    return 0

# TODO: get the node address where the loop begins
def has_cycle_floyd(head):
    if not head: return None
    fast = slow = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return 1
    return 0
```

```python
# hackerrank find mergenodes
# no optimization but this works even if lists do not merge
def findMergeNode(head1, head2):
    """
    prerequisites:
    head1!=head2
    head1 is not None and head2 is not None
    """
    while head1:
        cur = head2
        while cur:
            if cur.next == head1.next:
                return cur.next.data
            else:
                cur = cur.next
        head1 = head1.next
    return None

# this is more optimized but only works if it is given that lists always merge
def findMergeNodeHelp(head1, head2):
    """
    prerequisites:
    head1!=head2
    head1 is not None and head2 is not None
    """
    cur1 = head1
    cur2 = head2
    while cur1!=cur2:
        if not cur1.next:
            cur1 = head2
        else:
            cur1 = cur1.next
        if not cur2.next:
            cur2 = head1
        else:
            cur2 = cur2.next
    return cur2.data

# other people's solution
#same as findMergeNodeHelp but more pythonic
def FindMergeNode(headA, headB):
    a, b = headA, headB
    while a.data != b.data:
        a = a.next or headB
        b = b.next or headA
    return a.data
```
